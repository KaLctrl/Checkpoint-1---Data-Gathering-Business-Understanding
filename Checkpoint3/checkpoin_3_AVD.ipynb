## Suicides In India
https://www.kaggle.com/datasets/rajanand/suicides-in-india

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


file_path = "/content/Suicides in India 2001-2012.csv"
df = pd.read_csv(file_path)

```

# Informasi Struktur Dataset

```python
print("\nInformasi Dataset:")
print(df.info())
```

# Visualisasi: Tren Bunuh Diri dari Tahun ke Tahun

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x='Year', y='Total', data=df, marker='o', ci=None)
plt.title("Tren Jumlah Bunuh Diri di India per Tahun")
plt.xlabel("Tahun")
plt.ylabel("Jumlah Bunuh Diri")
plt.grid(True)
plt.show()
```

# Visualisasi: Perbandingan Bunuh Diri Berdasarkan Gender

```python
plt.figure(figsize=(8, 6))
sns.barplot(x='Gender', y='Total', data=df, estimator=sum, ci=None, palette='pastel')
plt.title("Perbandingan Jumlah Bunuh Diri Berdasarkan Gender")
plt.xlabel("Gender")
plt.ylabel("Jumlah Bunuh Diri")
plt.show()
```

# Visualisasi: Distribusi Usia dalam Kasus Bunuh Diri

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Age_group', y='Total', data=df, palette='muted')
plt.title("Distribusi Kasus Bunuh Diri Berdasarkan Kelompok Usia")
plt.xlabel("Kelompok Usia")
plt.ylabel("Jumlah Bunuh Diri")
plt.show()
```

# Heatmap Korelasi untuk Kolom Numerik

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.select_dtypes(include=[np.number]).corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Heatmap Korelasi")
plt.show()
```

# Insight dan Tindakan
Insight & Action:
1. Terlihat adanya tren peningkatan atau penurunan jumlah bunuh diri dari tahun ke tahun.
2. Kasus bunuh diri lebih banyak terjadi pada gender tertentu, sehingga kampanye pencegahan dapat difokuskan pada kelompok rentan.
3. Kelompok usia tertentu memiliki distribusi bunuh diri lebih tinggi, yang dapat menjadi target intervensi lebih lanjut.
4. Korelasi antara faktor-faktor tertentu dapat membantu mengidentifikasi penyebab utama kasus bunuh diri.
